### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
# Data for function y = x^2
x = np.linspace(-4, 4, 100)
y = x**2
```

```python
#Object-Oriented Approach
fig,ax, = plt.subplots()
ax.plot(x,y,color ='blue',label = 'y =x^2')
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()
plt.show()
```

```python
#Pyplot Approach
plt.figure()
plt.plot(x, y, color ='green', label = 'y = x^2')
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
#Create fig and subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize =(12 ,5))

#Scatter PLot
ax1.scatter(random_data_x, random_data_y, alpha = 0.5, color = 'purple')
ax1.set_xlabel("Random Data x")
ax1.set_ylabel("Random  Data Y")

#Bar plot
ax2.bar(names, values, color ='orange')
ax2.set_title("Bar plot of Fruits")
ax2.set_xlabel("Fruits")
ax2.set_ylabel("Count")

#show the figure with 2 subplots
plt.tight_layout()
plt.show()
```
